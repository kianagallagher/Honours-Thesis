```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
import os
```

```python
# gets the counts for each energy bin
def get_image_3x3_czt_ct(base_filename, bin, image_n):
  filearray = filearray = [
        ["M62638-A0.mat", "M62638-A1.mat"],
        ["M62639-A0.mat", "M62639-A1.mat"],
        ["M62641-A0.mat", "M62641-A1.mat"],
        ["M62646-A0.mat", "M62646-A1.mat"],
        ["M62651-A0.mat", "M62651-A1.mat"],
        ["M62652-A0.mat", "M62652-A1.mat"],
        ["M62653-A0.mat", "M62653-A1.mat"],
        ["M62654-A0.mat", "M62654-A1.mat"]
    ]

  filename = base_filename + filearray[0][0]
  mat_data = scipy.io.loadmat(filename) # loading in the air scan data

  data_in = mat_data["cc_struct"]
  data = data_in["data"][0,0]
  cc_data = data["cc_data"][0,0] # aquisition number, bin number, frame number, x, y

  nframes = cc_data.shape[2]
  rows = cc_data.shape[3]
  cols = cc_data.shape[4]

  output = np.zeros((nframes, 192, 72)) # 72 is the number of slices

  for i in range(len(filearray)):
    for j in range(len(filearray[0])):
      filename = base_filename + filearray[i][j]
      mat_data = scipy.io.loadmat(filename)

      data_in = mat_data["cc_struct"]
      data = data_in["data"][0,0]
      cc_data = data["cc_data"][0,0]

      for k in range(nframes):
        slice_2d = cc_data[image_n, bin, k, :, :]

        if j==0:
          slice_2d = np.fliplr(np.flipud(slice_2d))

        row_start = 24 * i
        row_end = row_start + 24
        col_start = 36 * j
        col_end = col_start + 36

        output[k, row_start:row_end, col_start:col_end] = slice_2d.astype(np.float64)

  return output
```

```python
FILENAME_BASE = '/content/gdrive/MyDrive/XCITE/Honours Project/Reconstruction/2_NEW_DETECTOR_RECON/2_DATA/S1728_ChickenDrumstick4_25_33_41_50_59_68_120kVp/' # source file
NUM_IMAGES = 6 # number of images from the source file
```

```python
os.makedirs(FILENAME_BASE+"Sinograms", exist_ok=True) # making a directory to save the results

for i in range(NUM_IMAGES):
  bin1_image = get_image_3x3_czt_ct(FILENAME_BASE, 0, i)
  bin2_image = get_image_3x3_czt_ct(FILENAME_BASE, 1, i)
  bin3_image = get_image_3x3_czt_ct(FILENAME_BASE, 2, i)
  bin4_image = get_image_3x3_czt_ct(FILENAME_BASE, 3, i)
  bin5_image = get_image_3x3_czt_ct(FILENAME_BASE, 4, i)
  bin6_image = get_image_3x3_czt_ct(FILENAME_BASE, 5, i)
  bin_all_image2 = get_image_3x3_czt_ct(FILENAME_BASE, 6, i)
  bin_all_image = bin1_image + bin2_image + bin3_image + bin4_image + bin5_image + bin6_image

  np.save(FILENAME_BASE+f"Sinograms_Summed/{chr(i+65)}_bin1_image.npy", bin1_image)
  np.save(FILENAME_BASE+f"Sinograms_Summed/{chr(i+65)}_bin2_image.npy", bin2_image)
  np.save(FILENAME_BASE+f"Sinograms_Summed/{chr(i+65)}_bin3_image.npy", bin3_image)
  np.save(FILENAME_BASE+f"Sinograms_Summed/{chr(i+65)}_bin4_image.npy", bin4_image)
  np.save(FILENAME_BASE+f"Sinograms_Summed/{chr(i+65)}_bin5_image.npy", bin5_image)
  np.save(FILENAME_BASE+f"Sinograms_Summed/{chr(i+65)}_bin6_image.npy", bin6_image)
  np.save(FILENAME_BASE+f"Sinograms_Summed/{chr(i+65)}_bin_all_image.npy", bin_all_image)
```
