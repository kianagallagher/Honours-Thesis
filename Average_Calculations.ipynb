```python
# 95% CI
def ci_95(avg, std, data):
    n = len(data)
    margin = 1.96 * (std / np.sqrt(n))
    lower = avg - margin
    upper = avg + margin
    lower_val = f"{lower:.9f}"
    upper_val = f"{upper:.9f}"
    return [lower_val, upper_val]
```
