The goal of this script is to mix the scans to create a train, validation, and test dataset of the reconstructed images.

```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
BASE_FOLDER = "/content/gdrive/MyDrive/Honours_Thesis/1.2_DATA/ChickenDrumstick/"
NEW_FOLDER = "360_to_45"
GT_FOLDER = "360"
RD_FOLDER = "45"
NUM_TOTAL_IMAGES = 24 # the number of scans
np.random.seed(25) # seed to make the randomness consistent
```

```python
# the approximate percentage to partition the data
TRAIN_PERCENT = 0.7
VALIDATION_PERCENT = 0.15
TEST_PERCENT = 0.15

TRAIN_NUM = np.ceil(NUM_TOTAL_IMAGES*TRAIN_PERCENT)
VALIDATION_NUM = np.floor(NUM_TOTAL_IMAGES*VALIDATION_PERCENT)
TEST_NUM = np.ceil(NUM_TOTAL_IMAGES*TEST_PERCENT)
```

```python
train = 0
test = 0
validate = 0

# making the directories to store the results
os.makedirs(f"{BASE_FOLDER}{NEW_FOLDER}")
os.makedirs(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}")

os.makedirs(f"{BASE_FOLDER}{NEW_FOLDER}/test/{GT_FOLDER}", exist_ok=True)
os.makedirs(f"{BASE_FOLDER}{NEW_FOLDER}/train/{GT_FOLDER}", exist_ok=True)
os.makedirs(f"{BASE_FOLDER}{NEW_FOLDER}/validate/{GT_FOLDER}", exist_ok=True)
os.makedirs(f"{BASE_FOLDER}{NEW_FOLDER}/test/{RD_FOLDER}", exist_ok=True)
os.makedirs(f"{BASE_FOLDER}{NEW_FOLDER}/train/{RD_FOLDER}", exist_ok=True)
os.makedirs(f"{BASE_FOLDER}{NEW_FOLDER}/validate/{RD_FOLDER}", exist_ok=True)

os.makedirs(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/test/{GT_FOLDER}", exist_ok = True)
os.makedirs(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/train/{GT_FOLDER}", exist_ok=True)
os.makedirs(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/validate/{GT_FOLDER}", exist_ok=True)
os.makedirs(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/test/{RD_FOLDER}", exist_ok=True)
os.makedirs(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/train/{RD_FOLDER}", exist_ok=True)
os.makedirs(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/validate/{RD_FOLDER}", exist_ok=True)

sorted_files = sorted(os.listdir(f"{BASE_FOLDER}{GT_FOLDER}_Projections_Images_512x512"))
image_num = 0

for file in sorted_files:
  # load in the ground truth images and the reduced dose images
  bin_images_gt = np.load(f"{BASE_FOLDER}{GT_FOLDER}_Projections_Images_512x512/{file}")
  bin_images_rd = np.load(f"{BASE_FOLDER}{RD_FOLDER}_Projections_Images_512x512/{file}")

  shuffle_index = np.arange(72)

  for k in range(72): # 72 slices
    slice_index = shuffle_index[k]
    bin1_gt, bin2_gt, bin3_gt, bin4_gt, bin5_gt, bin6_gt, binall_gt = bin_images_gt
    bin1_rd, bin2_rd, bin3_rd, bin4_rd, bin5_rd, bin6_rd, binall_rd = bin_images_rd

    bin1_rd = bin1_rd[slice_index, :, :]
    bin2_rd = bin2_rd[slice_index, :, :]
    bin3_rd = bin3_rd[slice_index, :, :]
    bin4_rd = bin4_rd[slice_index, :, :]
    bin5_rd = bin5_rd[slice_index, :, :]
    bin6_rd = bin6_rd[slice_index, :, :]
    binall_rd = binall_rd[slice_index, :, :]
    input_rd = np.stack([bin1_rd, bin2_rd, bin3_rd, bin4_rd, bin5_rd, bin6_rd], axis=0)

    bin1_gt = bin1_gt[slice_index, :, :]
    bin2_gt = bin2_gt[slice_index, :, :]
    bin3_gt = bin3_gt[slice_index, :, :]
    bin4_gt = bin4_gt[slice_index, :, :]
    bin5_gt = bin5_gt[slice_index, :, :]
    bin6_gt = bin6_gt[slice_index, :, :]
    binall_gt = binall_gt[slice_index, :, :]

    # save all the images from one scan to the same folder
    if (image_num < TRAIN_NUM):
        np.save(f"{BASE_FOLDER}{NEW_FOLDER}/train/{GT_FOLDER}/{train}.npy", binall_gt)
        np.save(f"{BASE_FOLDER}{NEW_FOLDER}/train/{RD_FOLDER}/{train}.npy", input_rd)

        np.save(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/train/{GT_FOLDER}/{train}.npy", binall_gt)
        np.save(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/train/{RD_FOLDER}/{train}.npy", binall_rd)
        train += 1

    elif (image_num < TRAIN_NUM+TEST_NUM):
        np.save(f"{BASE_FOLDER}{NEW_FOLDER}/test/{GT_FOLDER}/{test}.npy", binall_gt)
        np.save(f"{BASE_FOLDER}{NEW_FOLDER}/test/{RD_FOLDER}/{test}.npy", input_rd)

        np.save(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/test/{GT_FOLDER}/{test}.npy", binall_gt)
        np.save(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/test/{RD_FOLDER}/{test}.npy", binall_rd)
        test += 1

    else:
        np.save(f"{BASE_FOLDER}{NEW_FOLDER}/validate/{GT_FOLDER}/{validate}.npy", binall_gt)
        np.save(f"{BASE_FOLDER}{NEW_FOLDER}/validate/{RD_FOLDER}/{validate}.npy", input_rd)

        np.save(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/validate/{GT_FOLDER}/{validate}.npy", binall_gt)
        np.save(f"{BASE_FOLDER}Conventional_{NEW_FOLDER}/validate/{RD_FOLDER}/{validate}.npy", binall_rd)
        validate += 1

  image_num += 1 # increment the number of images saved
```

```python
# print out the results to make sure every thing matches
print(train)
print(test)
print(validate)
```
